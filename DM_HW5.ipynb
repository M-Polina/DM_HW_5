### b)
 ####  Find |V |, |E|, δ (G), Δ(G), rad(G), diam(G), girth(G), center(G), к(G), λ(G).
 ###### Solution:

```python
G = graph.subgraph(max(nx.connected_components(graph), key=len))
print("Answer: \n|V|:", graph.number_of_nodes())
print("|E|:", graph.number_of_edges())
min_graph_deg = 1000
max_graph_deg = 0

for country, deg in G.degree():
    max_graph_deg = max(max_graph_deg, deg)
    min_graph_deg = min(min_graph_deg, deg)
    
print('Minimal graph degree:', min_graph_deg)
print('Maximal graph degree:', max_graph_deg)

print('Radius:', nx.radius(G))
print('Diameter:', nx.diameter(G))
print('Girth:', min([len(cycle) for cycle in nx.cycle_basis(G)]))
print('Center:', nx.center(G))
      
print("\n к(G) Vertex connectivity:", nx.node_connectivity(G), \
          'for Minimum vertex cut:', nx.minimum_node_cut(G))
print("\nλ(G) Edge connectivity:", nx.edge_connectivity(G),\
          'for Minimum edge cut:', nx.minimum_edge_cut(G))
```

 ### c)
  ####  Find the minimum vertex coloring Z : V → N of G.
 ###### Solution:
 I used nx.greedy_color() to get the minimum vertex coloring
 I used Coloring algorithm, where I picked "smallest_last", as i need the minimum number of colors

```python
countries_colors = nx.coloring.greedy_color(G.subgraph(G), strategy="smallest_last")
print("Answer: \n minimum number of colors (size of minimum vertex coloring):", 1 + max(countries_colors.values()))

color_map = []
for country in G:
    color_map.append(countries_colors[country])

plt.figure(figsize=(14, 9))
nx.draw_planar(G, node_color=color_map, with_labels=True)
```

### l) 
  #### Find all biconnected components (blocks) and draw the block-cut tree of G*.
 ###### Solution:

```python
bi_comp = list(nx.biconnected_components(graph))

BC_tree = nx.Graph()

for point in nx.articulation_points(graph):
    for i in range(len(bi_comp)):
        if point in bi_comp[i]:
            BC_tree.add_edge(point, i + 1)

plt.figure(figsize=(8, 4))
print("Answer:")
nx.draw_planar(BC_tree, with_labels=True, node_size=1000)
```
